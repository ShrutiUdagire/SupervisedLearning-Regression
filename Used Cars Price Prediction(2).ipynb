```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_train, lin_reg.predict())
rmse=np.sqrt(mse)
rmse
```

# Assumptions of linear Regression

```python
# 1)No AutoCorrelation
import statsmodels.tsa.api as smt

acf = smt.graphics.plot_acf(lin_reg.resid, lags=40 , alpha=0.05)
acf.show()
```

As Durbin-Watson:1.991
    It Means No AutoCorrelation

```python
# 2)Homoskedaticity
```

```python
# Predicted VS Residuals
sns.regplot(x=lin_reg.predict(), y=lin_reg.resid, lowess=True, line_kws={'color': 'red'})
```

```python
# goldfeldquandt
from statsmodels.compat import lzip
import statsmodels.stats.api as sms
name = ['F statistic', 'p-value']
test = sms.het_goldfeldquandt(lin_reg.resid, lin_reg.model.exog)
lzip(name, test)
```

H0:Residuals are homoskedatic in nature


H1:Residuals are not homoskedatic in nature
    
Pvalue>0.05.


We fail to reject Null hypothesis

# DOUBT OF NORMALITY

```python
# 2)Normality
# DOUBT
name = ['Jarque-Bera', 'Chi^2 two-tail prob.', 'Skew', 'Kurtosis']
test = sms.jarque_bera(lin_reg.resid)
lzip(name, test)
```

```python
import seaborn as sns
sns.distplot(lin_reg.resid)
```

```python
import scipy.stats as stats
import pylab
from statsmodels.graphics.gofplots import ProbPlot
stats.probplot(lin_reg.resid, dist="norm", plot = pylab)
plt.show()
```

```python
# 4)Linearity
```

```python
sns.regplot(x=lin_reg.predict(), y=y_train, lowess=True,  line_kws={'color': 'red'})
```

```python
import statsmodels.api as sm
sm.stats.diagnostic.linear_rainbow(lin_reg, frac=0.5)
```

H0:Residuals are Linear
    
H1:residuals are not Linear
    
pvalue>0.05 

hence Fail to Reject Null Hypothesis

Hence Residuals are Linear

```python
# 5) No multicollineratity
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = [variance_inflation_factor(X_constant.values, i) for i in range(X_constant.shape[1])]
pd.DataFrame({'vif': vif[1:]}, index=X.columns).T
```

All Values are below 5.

Hence there is No multicollineratity
